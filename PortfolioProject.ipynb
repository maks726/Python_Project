```python
import matplotlib 
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib import figure

import seaborn as sns
import pandas as pd
import numpy as np

pd.options.mode.chained_assignment = None
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #size of plots

df = pd.read_csv(r'C:\Users\maxim\OneDrive\Рабочий стол\movies.csv') #read data

#pd.set_option('display.max_rows', None)
```

```python
df.head()
```

```python
for col in df: #missing data
    mis_data = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,round(mis_data*100)))
```

```python
df1 = df.dropna()
for col in df1: #missing data
    mis_data1 = np.mean(df1[col].isnull())
    print('{} - {}%'.format(col,round(mis_data1*100)))
```

```python
df1.dtypes
```

```python
df1['votes'] = df1['votes'].astype('int64')
```

```python
df1['budget'] = df1['budget'].astype('int64')
df1['gross']  = df1['gross'].astype('int64')
```

```python
df1['runtime']  = df1['runtime'].astype('int64')
```

```python
df1.boxplot(column=['gross'])  #outliers
```

```python
df1.drop_duplicates()
```

```python
df1.sort_values(by = ['score'], inplace = False, ascending = False)
```

```python
# budget vs IMDb rating
sns.regplot(x="budget", y="score", data=df1)
```

```python
# budget vs IMDb rating

sns.regplot(x = "votes", y = "score", data = df1)
```

```python
# correlation matrix
df1.corr(method ='pearson')
```

```python
df1.corr(method ='kendall')
```

```python
df1.corr(method ='spearman')
```

```python
correlation_matrix = df1.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix")

plt.show()
```

```python
# random numeric value for each unique categorical value

df1.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
correlation_matrix = df1.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix with numeric values")
plt.show()
```
